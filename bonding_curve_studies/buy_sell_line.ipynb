```python
import param as pm
import panel as pn


import hvplot.pandas
import pandas as pd
import numpy as np
import plotly
import plotly.express as px

pn.extension(plotly)
```

```python

class line_buy_sell_margin(pm.Parameterized):
    
    
    buy_adjustment = pm.Number(0, bounds=(-20, 20))
    sell_adjustment = pm.Number(0, bounds=(-20, 20))
    
    
    def __init__(self) -> None:
        super().__init__()
        
        
        
    def x_axis(self):
        x = np.linspace(0, 500, 501)
        return x
    
    
    def function_generator(self):

        
        # Making buy and sell functions
        buy_price_function = np.poly1d([2,410 + self.buy_adjustment])
        sell_price_function = np.poly1d([2,340 + self.sell_adjustment])
        
        return buy_price_function, sell_price_function
    
    def full_view(self):
        

        # Generating x axis
        x = self.x_axis()
    

        # Making y axes for both buy and sell

        buy_func, sell_func = self.function_generator()
        
        buy_y = buy_func(x)
        sell_y = sell_func(x)
        
        # Making the Dataframe for both the buy and sell functions
        combined_plot = pd.DataFrame(zip(x, buy_y, sell_y), columns=['supply','buy_price', 'sell_price'])
        
        plotly_element = px.line(combined_plot, x="supply", y=['buy_price','sell_price'])
        
        final_pane = pn.pane.Plotly(plotly_element, config={'responsive': True})
        return final_pane
    
        #return plotly_element
        
    def buy_area(self):
        
        x = self.x_axis()
        
        # Just grabbing the buy function
        buy_func, sell_func = self.function_generator()
        buy_y = buy_func(x)

        buy_plot = pd.DataFrame(zip(x, buy_y), columns=['supply','buy_price'])
        
        return px.area(buy_plot, x="supply", y=['buy_price'])

        
    def sell_area(self):
        
        x = self.x_axis()
        
        # Just grabbing the buy function
        buy_func, sell_func = self.function_generator()
        sell_y = sell_func(x)

        sell_plot = pd.DataFrame(zip(x, sell_y), columns=['supply','sell_price'])
        
        
        return px.area(sell_plot, x="supply", y=['sell_price'])
    
    
    def area_difference(self):
        return self.buy_area() * self.sell_area()
```

```python
x = line_buy_sell_margin()
x.full_view()
```
